```python
from sklearn.feature_selection import VarianceThreshold
import pandas as pd

# Assuming 'X' is your feature matrix (a pandas DataFrame or a NumPy array)
# Example DataFrame with constant and non-constant features:
data = {
    'feature1': [1, 1, 1, 1],
    'feature2': [2, 2, 2, 2],
    'feature3': [3, 3, 3, 4],
    'feature4': [4, 4, 4, 5]
}

X = pd.DataFrame(data)

# Instantiate VarianceThreshold with threshold=0
variance_threshold = VarianceThreshold(threshold=0)

# Fit the VarianceThreshold to your data
variance_threshold.fit(X)

# Get boolean mask indicating which features are retained
mask = variance_threshold.get_support()
print(mask)
# Get the constant and non-constant features
constant_features = X.columns[~mask]
print(constant_features)
non_constant_features = X.columns[mask]
print(non_constant_features)
# Display the results
print("Constant Features:", constant_features)
print("Non-Constant Features:", non_constant_features)
```

```python
number_of_columns = []
for i in df1:
    if(df1[i].dtypes != 'object'):
        number_of_columns.append(i)
```

```python
newdf = df1[number_of_columns]
```

```python
newdf.head()
```

```python
correlation_matrix = newdf.corr()
correlation_matrix
```

```python
threshold = 0.7
correlated_features = set()

for i in range(len(correlation_matrix.columns)):
    #print(i)
    for j in range(i):
        #print(j)
        if abs(correlation_matrix.iloc[i, j]) > threshold:
            #print(correlation_matrix.iloc[i,j])
            colname = correlation_matrix.columns[i]
            correlated_features.add(colname)
print("The highly correlated features are:", correlated_features)
```

```python
df1.drop(columns = correlated_features, inplace = True)
```
