**7. Calculate the 5 number summary and correlate with box plot (Python code)**

```python
df.shape
```

```python
df.drop_duplicates(inplace = True)
df.dropna(inplace = True)
```

```python
df.shape
```

```python
plt.boxplot(df['Quantity'])
plt.show()
```

```python
Q1 = df['Quantity'].quantile(0.25)
Q3 = df['Quantity'].quantile(0.75)
IQR = Q3 - Q1
```

```python
lb = Q1 - (IQR * 1.5)
ub = Q3 + (IQR * 1.5)
df = df[(df['Quantity'] > lb) & (df['Quantity'] < ub)]
```

```python
plt.boxplot(df['Quantity'])
plt.show()
```

**8. Display the outlier values using Z-score (Python code)**

```python
z_score = (numerical - numerical.mean()) / numerical.std()
outliers =  z_score[(z_score < -3) | (z_score > 3)]
outliers['Quantity']
```

**10. Analyze the skewness of the features using plot distribution graph and display the
features with right skew, left skew and no skew (Python code)**

```python
sns.kdeplot(numerical['Price'])
```

```python
skewness = numerical.skew()
```

```python
for i in skewness:
    if(i > 0):
        print("{0} is Right Skew".format(i))
    elif(i < 0):
        print("{0} is Left Skew".format(i))
    elif(i > -0.5) or (i < 0.5):
        print("{0} has No Skew".format(i))
```
